```python
#4 Construir una tabla que muestre cuántas medallas obtuvieron los hombres en total en cada año que se realizó el evento.
```

```python
men_medals = df_olympics.groupby(df_olympics.index).count()
print(men_medals)

```

```python
#Primero filtramos por hombre al dataframe
men_medals_1 = df_olympics.loc[df_olympics.Gender == 'Men']
men_medals_1
```

```python
#Ahora se crea un nuevo data frame agrupando los años, y contando cada fila. Mostrando el resultado para cada
men_medals_2 = men_medals_1.groupby(men_medals_1.index).count()
men_medals_3 = men_medals_2[['Medal']]
men_medals_3
```
