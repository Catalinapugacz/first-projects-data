En este link se provee un archivo con los resultados de la Encuesta de Sueldos de Openqube (https://sueldos.openqube.io/encuesta-sueldos-2020.02/), en este caso de Febrero 2020. Calcular y comparar media y mediana de los sueldos netos.


```python
import pandas as pd
```

```python
#Importando el archivo como csv
url = '/content/drive/MyDrive/Colab Notebooks/Copia de Salarios Openqube - 2020.02.csv'
```

```python
df = pd.read_csv(url)
print(df)
```

```python
#Seleccionando la columna que contiene los salarios
salarios = df['Salario mensual NETO (en tu moneda local)']

```

```python
print(salarios)
```

```python
#transformando en Array
salarios_array = np.array(salarios)
print(salarios_array)
print(type(salarios_array))
```

```python
listaSalario_array = []

for i in salarios_array:
  valor_string = str(i)
  valor_cambiado = valor_string.replace(".", "")
  listaSalario_array.append(float(valor_cambiado))
  
print(listaSalario_array)
```

```python
type(listaSalario_array)
```

```python
(listaSalario_array[1]) + (listaSalario_array[2])
```

```python
#Eliminando nans y nulls
newlist = [x for x in listaSalario_array if pd.isnull(x) == False and x != 'nan']
print(newlist)
```

```python
print(sum(newlist))
```

```python
mediana = np.median(newlist)
print(mediana)
```
